```python
from collections import deque
```

```python


def bfs(graph, start, goal=None):
    visited = set()
    queue = deque()
    
    # Start node
    visited.add(start)
    queue.append(start)
    
    while queue:
        node = queue.popleft()
        print(node, end=" ")
        
        # Goal check (optional)
        if goal is not None and node == goal:
            print("\nGoal Found!")
            return
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    if goal:
        print("\nGoal Not Found!")
```

```python
graph = {
    'A': ['S', 'B'],
    'S': ['G', 'C'],
    'B': [],
    'G': ['F', 'H'],
    'C': ['D', 'E'],
    'F': ['H'],
    'H': [],
    'D': [],
    'E': []
}

bfs(graph, 'A', 'E')
```

Shortest Path

```python
def bfs_path(graph, start, goal):
    visited = set()
    queue = deque([(start, [start])])
    
    while queue:
        node, path = queue.popleft()
        
        if node == goal:
            return path
        
        if node not in visited:
            visited.add(node)
            
            for neighbor in graph[node]:
                queue.append((neighbor, path + [neighbor]))

    return None

print("Shortest Path:", bfs_path(graph, 'A', 'E'))
```

BFS Pseudo Code (General Graph Traversal)

BFS(Graph, start):

1. Create an empty Queue Q
2. Create an empty Set Visited

3. Add start to Visited
4. Enqueue start into Q

5. While Q is not empty:

       a. current ← Dequeue(Q)
       b. Process current   // print or check goal

       c. For each neighbor in Graph[current]:

              If neighbor is NOT in Visited:
                     Add neighbor to Visited
                     Enqueue neighbor into Q


BFS with Goal Test (State Space Search Version)


BFS(Graph, start, goal):

1. Create Queue Q
2. Create Set Visited

3. Add start to Visited
4. Enqueue start into Q

5. While Q is not empty:

       a. current ← Dequeue(Q)

       b. If current == goal:
              Return "Goal Found"

       c. For each neighbor of current:

              If neighbor not in Visited:
                     Add neighbor to Visited
                     Enqueue neighbor

6. Return "Goal Not Found"


BFS for Shortest Path

BFS_Path(Graph, start, goal):

1. Create Queue Q
2. Enqueue (start, [start])   // store path also
3. Create Set Visited

4. While Q is not empty:

       a. (current, path) ← Dequeue(Q)

       b. If current == goal:
              Return path

       c. If current not in Visited:
              Add current to Visited

              For each neighbor of current:
                     Enqueue (neighbor, path + neighbor)

5. Return "No Path"
