DFS using Recursion (Most Common)

```python
def dfs(graph, node, visited):
    visited.add(node)          # mark node as visited
    print(node, end=" ")       # process the node
    
    for neighbour in graph[node]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)


# Example graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

visited = set()
dfs(graph, 'A', visited)
```

DFS using Stack (Iterative Version)

```python
# DFS using Stack (Iterative)

def dfs_iterative(graph, start):
    visited = set()
    stack = [start]   # Initialize stack with start node
    
    print("DFS Traversal:")
    
    while stack:
        node = stack.pop()   # LIFO (Last In First Out)
        
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            
            # Add neighbours to stack (reverse for correct order)
            for neighbour in reversed(graph[node]):
                if neighbour not in visited:
                    stack.append(neighbour)


# Example Graph (Adjacency List)
graph = {
    0: [1, 2],
    1: [3, 4],
    2: [5],
    3: [],
    4: [],
    5: []
}

# Call DFS
dfs_iterative(graph, 0)
```

DFS to Find Goal Node (Iterative â€“ Stack Based)

```python
def dfs_find_goal(graph, start, goal):
    visited = set()
    stack = [start]   # Initialize stack
    
    while stack:
        node = stack.pop()   # LIFO
        
        if node not in visited:
            print("Visiting:", node)
            visited.add(node)
            
            # Check if goal found
            if node == goal:
                print("Goal node found!")
                return True
            
            # Add neighbours to stack
            for neighbour in reversed(graph[node]):
                if neighbour not in visited:
                    stack.append(neighbour)
    
    print("Goal node not found.")
    return False


# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# Search for goal
dfs_find_goal(graph, 'A', 'E')
```
